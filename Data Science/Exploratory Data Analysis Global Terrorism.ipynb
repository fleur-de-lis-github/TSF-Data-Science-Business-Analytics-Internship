# Data Preprocessing

```python
dataset.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','provstate':'states','region_txt':'Region','country_txt': 'Country',
                      'attacktype1_txt':'AttackType','nkill':'Killed','target1':'Target','nwound':'Wounded','gname':'GroupName',
                      'targtype1_txt':'Target_type','weaptype1_txt':'Weapontype','city':'City','motive':'Motive'},inplace =True)
```

```python
dataset.head(5)
```

# Finding Number of Attacks Each Year

```python
year = dataset['Year'].unique()
years_count = dataset['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))

sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=30)

plt.show()
```

# Analyzing Type of Attacks

```python
dataset["AttackType"].value_counts().plot(kind='bar',figsize=(20,10),color='teal')

plt.xticks(rotation = 50)
plt.xlabel("AttackType",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Name of attacktype",fontsize=20)

plt.show()
```

# Number of Deaths

```python
df= dataset[['Year','Killed']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)

plt.xticks(rotation = 50)
plt.title("People Died Due To Attack",fontsize=25)
plt.ylabel("Number of Deaths",fontsize=20)
plt.xlabel('Year',fontsize=20)

top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
dataset[['AttackType','Wounded']].groupby(["AttackType"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['darkslateblue'])

plt.xticks(rotation=50)
plt.title("Number of wounded  ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack Type',fontsize=15)

plt.show()
```

# Number of Attacks Per Year

```python
plt.subplots(figsize=(20,10))

sns.countplot(dataset["Target_type"],order=dataset['Target_type'].value_counts().index,palette="gist_heat",edgecolor=sns.color_palette("mako"));

plt.xticks(rotation=90)
plt.xlabel("AttackType",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)

plt.show()
```

# Terrorist Groups with Highest Attack Count

```python
dataset['GroupName'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='green',figsize=(20,10))

plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("Attack number",fontsize=15)

plt.show()
```

# Terrorist Activities By Region

```python
pd.crosstab(dataset.Year, dataset.Region).plot(kind='area',stacked=False,figsize=(20,10))

plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)

plt.show()
```

# Top Countries Affected By Terrorism

```python
plt.subplots(figsize=(20,10))

sns.barplot(dataset['Country'].value_counts()[:10].index,dataset['Country'].value_counts()[:10].values,palette='YlOrBr_r')

plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 50)

plt.show()
```

# Top Affected Cities

```python
dataset['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='lightseagreen')

plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected city",fontsize=20)

plt.show()
```
